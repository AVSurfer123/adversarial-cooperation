```python
from pettingzoo.butterfly import knights_archers_zombies_v2 as kaz
env = kaz.env()
env.reset()
print(dir(env))
```

```python
print(env.action_spaces)
print(env.agent_selection)
print(env._agent_idxs)
print(env.observation_spaces)
```

```python
import imageio

vid_name = './kaz_knight0.mp4'
writer = imageio.get_writer(vid_name, fps=20)
```

```python
observation = env.reset()
action_spaces = env.action_spaces
count = 0
for agent in env.agent_iter(max_iter=10000):
    reward, done, info = env.last()
    action = action_spaces[env.agent_selection].sample()
    observation = env.step(action)
    if env.agent_selection == 'knight_0':
        writer.append_data(observation)
        count += 1
print(count)
writer.close()
```
