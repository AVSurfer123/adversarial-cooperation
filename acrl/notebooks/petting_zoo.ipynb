```python
import numpy as np
import matplotlib.pyplot as plt
import pettingzoo
import gym

%matplotlib inline
```

```python
from pettingzoo.butterfly import knights_archers_zombies_v2 as kaz, cooperative_pong_v1 as pong, pistonball_v0
env = pistonball_v0.env(local_ratio=0.5)
env.reset()
print(dir(env))
```

```python
print(env.action_spaces)
print(env.agent_selection)
print(env._agent_idxs)
print(env.observation_spaces)
```

```python
import imageio

vid_name = './pistonball.mp4'
writer = imageio.get_writer(vid_name, fps=20)
```

```python
observation = env.reset()
action_spaces = env.action_spaces
count = 0
total_reward = 0
iterations = 0
for agent in env.agent_iter(max_iter=10000):
    reward, done, info = env.last()
    if done:
        observation = env.reset()
        iterations += 1
    total_reward += reward
    action = action_spaces[agent].sample()
    observation = env.step(action)
    if env.agent_selection == 'piston_10':
        writer.append_data(observation)
        count += 1
print(count)
print("Average reward:", total_reward / iterations)
writer.close()
```
